```python
import chardet
from bs4 import BeautifulSoup

def detect_encoding(file_path):
    """Detect the file encoding using chardet."""
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']
```

```python
def extract_text_from_sgml(file_path):
    """Extract text from an SGML file."""
    encoding = detect_encoding(file_path)
    with open(file_path, 'r', encoding=encoding, errors='ignore') as f:
        sgml_content = f.read()
    soup = BeautifulSoup(sgml_content, 'html.parser')
    return soup.get_text()
```

Updating the leaflet dataset with "doctor's" answers summarizing the information from the document.

```python
# The function to get answers.
def get_doctors_answer(question):
    try:
        response = client.chat.completions.create(
            model='gpt-4o-mini',
            messages=[
                {
                    "role": "user",
                    "content": "Summarize the information from the leaflet and provide the most useful information for the patient about getting the medicine including its name, uses, 3 important side effects, and recommendations for taking the medicine. Here is the text: " + question
                }
            ],
            max_tokens=300
        )
        answer = response.choices[0].message.content
        return answer
#        return response['choices'][0]['message']['content']
#        return response.choices[0].message['content']
    except Exception as e:
        print(f"Error: {e}")
        return "Error in generating summary."
```

```python
# Iterate the document to update the dataframe with the answers.
for index, row in df_leaflets.iterrows():
    question = row['Patient']
    answer = get_doctors_answer(question)
    df_leaflets.at[index, 'Doctor'] = answer
```
